```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August',
          'September', 'October', 'November', 'December']
names = ['Temp', 'Prec', 'Direction', 'Speed', 'Temp_N', 'Prec_N', 'Direction_N',
         'Speed_N', 'Year', 'Month']
```

```python
def prepare_df(csv_name):
    with open(csv_name, 'r') as csv_file:
        with open('prepared.csv', 'w') as out_file:
            for line in csv_file:
                if 'Ш' in line:
                    new_list = []
                    list_line = line.split(',')
                    for el in list_line:
                        if el != 'Ш':
                            new_list.append(el)
                        else:
                            new_list.append('Ш')
                            new_list.append('0')
                    out_file.write(','.join(new_list))
                else:
                    out_file.write(line)

prepare_df('Poltava.csv')
```

```python
df = pd.read_csv('prepared.csv', header=None, names=names, usecols=range(1, 11))
df = df[pd.notnull(df.Speed_N)]
df['Speed'] = df['Speed'].str.replace(r'\D+', '').astype(int)
df['Speed_N'] = df['Speed_N'].str.replace(r'\D+', '').astype(int)
df['Year'] = df.Year.astype(int)
df['Month'] = df.Month.astype(int)
df['avg_temp'] = (df.Temp + df.Temp_N) / 2
df.info()
```

```python
avg_tmp_by_year = df.groupby('Year').avg_temp.mean()
avg_tmp_by_year.plot()
```

```python
avg_prec_by_year = df.groupby('Year').Prec.mean()
avg_prec_by_year.plot()
```
