```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August',
          'September', 'October', 'November', 'December']
names = ['Temp', 'Prec', 'Direction', 'Speed', 'Temp_N', 'Prec_N', 'Direction_N',
         'Speed_N', 'Year', 'Month']
```

```python
def prepare_df(csv_name):
    with open(csv_name, 'r') as csv_file:
        with open('prepared.csv', 'w') as out_file:
            for line in csv_file:
                if 'Ш' in line:
                    new_list = []
                    list_line = line.split(',')
                    for el in list_line:
                        if el != 'Ш':
                            new_list.append(el)
                        else:
                            new_list.append('Ш')
                            new_list.append('0')
                    out_file.write(','.join(new_list))
                else:
                    out_file.write(line)

prepare_df('Poltava.csv')
```
